```python
# import relevant packages including the functions and menus spyder files
import sqlite3
import pandas as pd
import sys
import functions as f
import menus as m


# set up the cursor
with sqlite3.connect("bike_sharing_system.db") as db:
    cursor = db.cursor()

# set it so that a max of 100 rows are displayed
pd.set_option('display.max_rows', 100)

# define a main function to run the whole system
def main():
    # first display the welcome menu to sign in/up
    m.welcome_menu()
    
    # check if the user is a customer
    customer = f.check_customer()
    # if user is a customer take them to customer menu
    if customer:
        m.customer_menu()
        # once they sign out restart the main menu
        main()
        
    # else check if the user is an admin    
    else:
        admin = f.check_admin()
        
        # if user is an admin check if they are an operator or a manager
        if admin:
            designation = f.get_designation()
            
            # if user is an operator take them to operator menu
            if designation == 'O':
                m.operator_menu()
                # once they sign out restart the main menu
                main()
                
            # else user is a manager so take them to manager menu    
            else:
                m.manager_menu()
                # once they sign out restart the main menu
                main()
```

```python
# run the main function
main()
```

```python
# close the database
db.close()
```
